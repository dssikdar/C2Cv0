```python
remove = ['Sst Chodl']
barcode_list = []
for ii,name in enumerate(data.obs_names):
    if data.obs.subclass[ii] in remove:
        barcode_list.append(False)
    else:
        barcode_list.append(True)
data_subset = data[barcode_list].copy()
```

```python
data = MultimodalData(data_subset)
```
