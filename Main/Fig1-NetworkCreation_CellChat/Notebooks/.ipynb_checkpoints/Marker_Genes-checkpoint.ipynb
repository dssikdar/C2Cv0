```r
cellchat_CON <- readRDS(file = '/extra/zhanglab0/CommonData/AMP-AD/Obj-CellChat/subclass_cellchat_control.rds')
cellchat_AD <- readRDS(file = '/extra/zhanglab0/CommonData/AMP-AD/Obj-CellChat/subclass_cellchat_alz.rds')
```

```r
data <- rjson::fromJSON(file="/srv/disk00/cellTypeAnnotation_pipeline/AHBA_PFC_filtered.json")
```

```r
celltype_data <- data$cell_types[[1]]$subtypes$cell_types
celltype_data2 <- data$cell_types[[2]]$subtypes$cell_types
```

```r
library(tidyverse)
```

```r
gene_list <- c()
cell_list <- c()
no_list <- c('CNTN5', 'ADARB2', 'GRIK1', 'FSTL5', 'GALNTL6', 'FGF13', 
             'IL1RAPL2', 'CADPS2', 'CBLN2', 'CPNE4', 'SST', 'TOX', 
            'NXPH1', 'TLE4', 'HS3ST4', 'MARCH1', 'PDZRN4', 'RMST',
            'MEIS2', 'SEMA3E', 'RELN', 'TRPM3', 'TSHZ2', 'POU6F2',
             'FOXP2', 'RORB', 'RXFP1', 'DCC', 'PTPRD', 'CXCL14')
# c('CNTN5', 'ADARB2', 'GRIK1', 'FSTL5', 'GALNTL6', 'FGF13', 
#              'IL1RAPL2', 'CADPS2', 'CBLN2', 'CPNE4', 'SST', 'TOX', 
#             'NXPH1', 'TLE4', 'HS3ST4', 'MARCH1', 'PDZRN4', 'RMST',
#             'MEIS2', 'SEMA3E', 'RELN', 'TRPM3', 'TSHZ2', 'POU6F2',
#             'PTPRK', 'FOXP2', 'SLIT3', 'RORB', 'CA10')
for(i in 1:length(celltype_data)){
    # print(i)
    name <- celltype_data[[i]]$name
    if (celltype_data[[i]]$name == 'L5 IT') {
        # print('L5 IT')
        name <- 'L5'
    }
    if (celltype_data[[i]]$name == 'L5 ET') {
        next
    }
    cell_list <- append(cell_list, glue(name, '_ALZ'))
    cell_list <- append(cell_list, glue(name, '_CON'))
    # print(i)
    # gene_list <- c()
    counter <- 0
    for (j in 1:length(celltype_data[[i]]$markers[[1]]$genes)) {
        if (counter == 2) {
            break
        }
        temp <-  substring(celltype_data[[i]]$markers[[1]]$genes[j], 1, str_length(celltype_data[[i]]$markers[[1]]$genes[j])-1)
        if (temp %in% gene_list) {
            next
        }
        if (temp %in% no_list){
            next
        }
        if (temp %in% rownames(cellchat_CON@data)) {
            gene_list <- append(gene_list, temp)
            counter = counter + 1
        }
    }
}

for(i in 1:length(celltype_data2)){
    # print(i)
    name <- celltype_data2[[i]]$name
    if (celltype_data2[[i]]$name == 'L5 IT') {
        # print('L5 IT')
        name <- 'L5'
    }
    if (celltype_data2[[i]]$name == 'L5 ET') {
        next
    }
    cell_list <- append(cell_list, glue(name, '_ALZ'))
    cell_list <- append(cell_list, glue(name, '_CON'))
    # gene_list <- c()
    counter <- 0
    for (j in 1:length(celltype_data2[[i]]$markers[[1]]$genes)) {
        if (counter == 2) {
            break
        }
        temp <-  substring(celltype_data2[[i]]$markers[[1]]$genes[j], 1, str_length(celltype_data2[[i]]$markers[[1]]$genes[j])-1)
        if (temp %in% gene_list) {
            next
        }
        if (temp %in% no_list){
            next
        }
        if (temp %in% rownames(cellchat_CON@data)) {
            gene_list <- append(gene_list, temp)
            counter = counter + 1
        }
    }
}
length(gene_list)
```

```r
# 32 x 32
df <- data.frame(matrix(NA, nrow = 34, ncol = 34))
# colnames(df) <- gene_list
# rownames(df) <- cell_list
colnames(df) <- cell_list
rownames(df) <- gene_list
df
```

```r
for(i in 1:length(celltype_data)){
    # print(i)
    name <- celltype_data[[i]]$name
    if (celltype_data[[i]]$name == 'L5 IT') {
        # print('L5 IT')
        name <- 'L5'
    }
    if (celltype_data[[i]]$name == 'L5 ET') {
        next
    }
    # print(i)
    df[,glue(name, '_CON')] <- rowMeans(cellchat_CON@data[gene_list,c(cellchat_CON@meta['subclass'] == name)])
    

    df[,glue(name, '_ALZ')] <- rowMeans(cellchat_AD@data[gene_list,c(cellchat_AD@meta['subclass'] == name)])
}

for(i in 1:length(celltype_data2)){
    # print(i)
    name <- celltype_data2[[i]]$name
    if (celltype_data2[[i]]$name == 'L5 IT') {
        # print('L5 IT')
        name <- 'L5'
    }
    if (celltype_data2[[i]]$name == 'L5 ET') {
        next
    }

    df[,glue(name, '_CON')] <- rowMeans(cellchat_CON@data[gene_list,c(cellchat_CON@meta['subclass'] == name)])

    df[,glue(name, '_ALZ')] <- rowMeans(cellchat_AD@data[gene_list,c(cellchat_AD@meta['subclass'] == name)])
}
```
